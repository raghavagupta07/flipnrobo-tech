```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.kayak.co.in/flights/BLR-DEL/2021-11-26?sort=duration_a"
driver.get(url)
time.sleep(6)
```

```python
#name
name = []

nam = driver.find_elements_by_xpath("//span[@class= 'codeshares-airline-names']")
for i in nam:
    if i.text is None:
        name.append('NAN')
    else:
        name.append(i.text)
name
```

```python
dep_time = []

dep = driver.find_elements_by_xpath("//div[@class= 'resultWrapper']/div[2]/div/div[1]/div[2]/div/ol/li/div/div/div[2]/div/span[1]/span")
for i in dep:
    if i.text is None:
        dep_time.append('NAN')
    else:
        dep_time.append(i.text)
        
arr_time = []

arr = driver.find_elements_by_xpath("//div[@class= 'resultWrapper']/div[2]/div/div[1]/div[2]/div/ol/li/div/div/div[2]/div/span[3]/span")
for i in arr:
    if i.text is None:
        arr_time.append('NAN')
    else:
        arr_time.append(i.text)
    
source = []

dep = driver.find_elements_by_xpath("//div[@class= 'resultWrapper']/div[2]/div/div[1]/div[2]/div/ol/li/div/div/div[2]/div[2]/div[1]/span")
for i in dep:
    if i.text is None:
        source.append('NAN')
    else:
        source.append(i.text)
        
destination = []

ar = driver.find_elements_by_xpath("//div[@class= 'resultWrapper']/div[2]/div/div[1]/div[2]/div/ol/li/div/div/div[2]/div[2]/div[2]/span")
for i in ar:
    if i.text is None:
        destination.append('NAN')
    else:
        destination.append(i.text)
        
stops = []

st = driver.find_elements_by_xpath("//div[@class= 'resultWrapper']/div[2]/div/div[1]/div[2]/div/ol/li/div/div/div[3]/div[1]/span[1]")
for i in st:
    if i.text is None:
        stops.append('NAN')
    else:
        stops.append(i.text)
        
duration = []

du = driver.find_elements_by_xpath("//div[@class= 'resultWrapper']/div[2]/div/div[1]/div[2]/div/ol/li/div/div/div[4]/div[1]")
for i in du:
    if i.text is None:
        duration.append('NAN')
    else:
        duration.append(i.text)
        
price = []

pri = driver.find_elements_by_xpath("//div[@class= 'resultWrapper']/div[2]/div/div[2]/div/div[2]/div[1]/div/div/div/a/span/span")
for i in pri:
    if i.text is None:
        price.append('NAN')
    else:
        price.append(i.text)
```

```python
print(len(name),len(dep_time),len(arr_time),len(source),len(destination),len(duration),len(stops),len(price))
```
