1. Write a python program which searches all the product under a particular product vertical
from www.amazon.in. The product verticals to be searched will be taken as input from user.
For e.g. If user input is ‘guitar’. Then search for guitars.
2. In the above question, now scrape the following details of each product listed in first 3 pages 
of your search results and save it in a data frame and csv. In case if any product vertical has 
less than 3 pages in search results then scrape all the products available under that product 
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of 
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" 
and “Product URL”. In case, if any of the details are missing for any of the product then 
replace it by “-“

```python
#finding urls
urls = []
for i in driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]'):
    urls.append(i.get_attribute("href"))
urls
```

```python
#defining varaibles
product=[]
brand = []
rating = []
price = []
exchange = []
deliver = []
available = []
no_Rating = []
Details = []
```

```python
for i in urls:
    driver.get(i)
    time.sleep(3)
    #fetching brand
    try:
        brands= driver.find_element_by_xpath('//a[@id="bylineInfo"]')
        brand.append(brands.text.replace('/n','new line'))
    except NoSuchElementException:
        brand.append('-')
        
     #fetching product
    try:
        products= driver.find_element_by_xpath('//h1[@id="title"]')
        product.append(products.text.replace('/n','new line'))
    except NoSuchElementException:
        product.append('-')
        
     #fetching Ratings
    try:
        ratings= driver.find_element_by_xpath('//span[@class="a-size-medium a-color-base"]')
        rating.append(ratings.text.replace('/n','new line'))
    except NoSuchElementException:
        rating.append('-')   
     
     #fetching prices
    try:
        prices= driver.find_element_by_xpath('//td[@class="a-span12"]')
        price.append(prices.text.replace('/n','new line'))
    except NoSuchElementException:
        price.append('-') 
        
     #fetching returns
    try:
        exchanges = driver.find_element_by_xpath('//div[@id="RETURNS_POLICY"]')
        exchange.append(exchanges.text.replace('/n','new line'))
    except NoSuchElementException:
        exchange.append('-')         
        
     #fetching Expected_Deliver
    try:
        Delivery= driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage"]')
        deliver.append(Delivery.text.replace('/n','new line'))
    except NoSuchElementException:
        deliver.append('-')
        
     #fetching Availability
    try:
        availables= driver.find_element_by_xpath('//div[@id="availability"]')
        available.append(availables.text.replace('/n','new line'))
    except NoSuchElementException:
        available.append('-')  
          
   #fetching Other_no.of.ratings     
    try:
        no_Ratings= driver.find_element_by_xpath('//span[@class="a-row a-spacing-medium averageStarRatingNumerical"]')
        no_Rating.append(no_Ratings.text.replace('/n','new line'))
    except NoSuchElementException:
        no_Rating.append('-')         
        
    
  #fetching Other_Detail
    try:
        Detail = driver.find_element_by_xpath('//div[@class="a-row a-spacing-top-base"]')
        Details.append(Detail.text.replace('/n','new line'))
    except NoSuchElementException:
        Details.append('-')
        
        
```

6. Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 –
September 20) from trak.in.

```python
#lets now import all required libraries 
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException  #importing exception handling library
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url='https://trak.in/'
driver.get(url)
```

```python
button = driver.find_element_by_xpath('//li[@id="menu-item-51510"]/a').get_attribute('href')
driver.get(button)
```

```python
Date = []
Startup_Name = []
Industry_Vertical = []
Sub_Vertical = []
Location = []
Investor = []
Investment_Type = []
Amount_in_usd = []
```

```python
for i in range(48,51):
    driver.find_element_by_xpath('//div[@id="tablepress-{}_wrapper"]/div/label/select/option[4]'.format(i)).click()

    # Date
    dt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for d in dt:
        Date.append(d.text)
 # Startup Name
    sn = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for n in sn:
        Startup_Name.append(n.text)
    
    # Industry_Vertical
    ind = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for n in ind:
        Industry_Vertical.append(n.text)
    
    # Sub_Vertical
    sv = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for s in sv:
        Sub_Vertical.append(s.text)

    # Location
    loc = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for l in loc:
        Location.append(l.text)
    
    # Investor
    inv = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[7]'.format(i))
    for n in inv:
        Investor.append(n.text)
    
    # Investment_Type
    invt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for n in invt:
        Investment_Type.append(n.text)
    
    # Amount_in_usd
    amt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for a in amt:
        Amount_in_usd.append(a.text)
```

```python
print(Date)
print(Startup_Name) 
print(Industry_Vertical) 
print(Sub_Vertical) 
print(Location) 
print(Investor)  
print(Investment_Type)
print(Amount_in_usd) 
```

```python
import pandas as pd
Funding_Deals=pd.DataFrame({})
Funding_Deals['Date']=Date
Funding_Deals['Startup Name']=Startup_Name
Funding_Deals['Location']=Location
Funding_Deals['Industry Vertical']=Industry_Vertical
Funding_Deals['Sub Vertical']=Sub_Vertical
Funding_Deals['Investor']=Investor
Funding_Deals['Investment Type']=Investment_Type
Funding_Deals['Amount In_USD']=Amount_in_usd
```

10. Write a python program to scrape a data for all available Hostels from 
https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, 
distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms 
from price, facilities and property description.

```python
#lets now import all required libraries 
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException  #importing exception handling library
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url='https://www.hostelworld.com/'
driver.get(url)
```

```python
loc= driver.find_element_by_id("search-input-field")
loc.send_keys("London")
```

```python
search_btn=driver.find_element_by_xpath("/html/body/div[1]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div")
search_btn.click()
```

```python
search_btn1=driver.find_element_by_id("search-button")
search_btn1.click()
```

```python
names=driver.find_elements_by_xpath("//h2[@class='title title-6']")
name=[]
for i in names:
    name.append(i.text)
name    
```

```python
dist=driver.find_elements_by_xpath("//span[@class='description']")
distnc=[]
for i in dist:
    distnc.append(i.text)
distnc
```

```python
len(distnc)
```

```python
ovralrev=driver.find_elements_by_xpath("//div[@class='keyword']")
overallrev=[]
for i in ovralrev:
    overallrev.append(i.text)
overallrev
```

```python
len(overallrev)
```

```python
totalrev = driver.find_elements_by_xpath("//div[@class='reviews']")
total_reviews=[]
for i in totalrev:
    total_reviews.append(i.text)
total_reviews
```

```python
len(total_reviews)
```

```python
rtngs=driver.find_elements_by_xpath("//div[@class='score orange big']")
ratings=[]
for i in rtngs:
    ratings.append(i.text)
```

```python
ratings
```

```python
len(ratings)
```

```python
dorms_pvt_price=driver.find_elements_by_xpath("//div[@class='price-col']")
dorms_pvt_prices=[]
for i in dorms_pvt_price:
    dorms_pvt_prices.append(i.text.replace('\n',''))
dorms_pvt_prices
```

```python
len(dorms_pvt_prices)
```

```python
pvt = []
for i in range(0,len(dorms_pvt_prices),2):
    pvt.append(dorms_pvt_prices[i])
pvt
```

```python
len(pvt)
```

```python
dorms = []
for i in range(1,len(dorms_pvt_prices),2):
    dorms.append(dorms_pvt_prices[i])
dorms
```

```python
len(dorms)
```

```python
facilitiess=driver.find_elements_by_xpath("//div[@class='has-wifi']")
facilityy=[]
for i in facilitiess:
    facilityy.append(i.text)
facilityy
```

```python
pptdesc=driver.find_elements_by_xpath("//div[@class='cleanliness']")
prptdesc=[]
for i in pptdesc:
    prptdesc.append(i.text)
prptdesc
```

```python
print(len(name[:27]),len(distnc[:27]),len(overallrev[:27]),len(total_reviews[:27]),len(ratings[:27]),len(pvt[:27]),len(dorms[:27]),len(facilityy[:27]),len(prptdesc[:27]))
```

```python
hostels=pd.DataFrame()
hostels["Hostel_Name"]=name[:27]
hostels["City_Distance"]=distnc[:27]
hostels["Overall_Review"]=overallrev[:27]
hostels["Total_Reviews"]=total_reviews[:27]
hostels["ratings"]=ratings[:27]
hostels["Privates_From_Price"]=pvt[:27]
hostels["Dorms_From_Price"]=dorms[:27]
hostels["facilities"]=facilityy[:27]
hostels["property description"]=prptdesc[:27]
```

```python
hostels
```

```python
#dataframe save it in csv file
hostels.to_csv('Hostel_Rooms.csv')
```
